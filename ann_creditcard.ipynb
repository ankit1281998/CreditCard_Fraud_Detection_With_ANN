```python
import pandas as pd
import numpy as np
import os
```

```python
pwd=os.getcwd()
data=pd.read_csv(pwd+"//creditcard.csv")
data
```

```python
data["Class"].value_counts()
```

```python
pd.value_counts(data["Class"]).plot.bar()
```

```python
X=data.iloc[:,1:30]
y=data["Class"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
```

```python
X_train
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
model=Sequential()
model.add(Dense(50,activation="relu",kernel_initializer="he_uniform",input_shape=(29,)))
model.add(Dense(1,activation="sigmoid"))
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.fit(X_train,y_train,epochs=10)
```

```python
y_pred1=model.predict(X_test)
```

```python
y_pred1
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,y_pred1)
```

## Weighted NN with Keras

```python
weights_assigned={0:1,1:600}
```

```python
model1=Sequential()
model1.add(Dense(50,activation="relu",kernel_initializer="he_uniform",input_shape=(29,)))
model1.add(Dense(1,activation="sigmoid"))
model1.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model1.fit(X_train,y_train,class_weight=weights_assigned,epochs=10)
```

```python
y_pred2=model1.predict(X_test)
y_pred2
```

```python
roc_auc_score(y_test,y_pred2)
```
